```python
X = df.drop('price',axis=1).values#meghdare hame be joz price
y = df['price'].values #meghdare faghat price
#inja moheme k values bashe chon tensorflow nmitune ba panda kar kone va bayad numpy bashe
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
#fit mikone be data train meghdare monasebe scale va transform mikone
```

```python
X_test = scaler.transform(X_test)#faghat transform mikone scale ro
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape #tedade satr o sotun pas mishe 19 featue o 19 neuron
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam',loss = 'mse')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),batch_size=128,epochs=400)
#bade har epoch sari test data ro mizane check mikone vali weight o bias ro taghyir nmide
#batch_size chon data bozorge daste daste midim
#harchi batch kuchiktar bashe zaman bishtare vali shanse overfit kamtare
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
#chon validation ham dare miad payin be ma mige k mitunnim train ro edame bedim 
#bdune tars az overfiting(nabayad val_loss biad bala)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
#bayad bozorg bashe chon gheymate khunast
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
#jazresh
```

```python
mean_absolute_error(y_test,predictions)
#miangin khata gheymat
```
