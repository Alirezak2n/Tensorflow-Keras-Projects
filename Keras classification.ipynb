```python
X = df.drop('benign_0__mal_1',axis = 1).values
y = df['benign_0__mal_1'].values#moshakhas karande feature ha o natije
```

```python
from sklearn.model_selection import train_test_split #split ro miarim
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
#anjam midim b in surat k x o y ro moshakhas karim in miad ghesmat mikone
```

```python
from sklearn.preprocessing import  MinMaxScaler #scaler ro miarim
```

```python
scaler=MinMaxScaler()#ye esme rahat midim
```

```python
X_train= scaler.fit_transform(X_train)#data test o train ro scale o transform karde
```

```python
X_test= scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential #model ro miarim
```

```python
from tensorflow.keras.layers import Dense,Dropout #layers ro miarim
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#layer akhar ro bayad ye activation bzarim k 0 o 1 bde mesle sigmoid
#chon k binary classification darim k khoruji 0 o 1 hast
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
#loss bayad in balayi bashe
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
losses#losses ro check mikonim o mibinim
```

```python
losses.plot()
#neshun mide k overfit shode chon val_loss miad bala 
```

```python
from tensorflow.keras.callbacks import EarlyStopping
#mizare ma model ro vaisunim age overfit shod
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
#chiro monitor kone va mode mige mikhaym be min beresim ya max
#masalan age loss ma 0 beshe baraye ma kheili khube
#verbose yani ye meghdar etelaat bde age 1 bashe age 0 yani nade
#patience yani ta 25 epoch vaisa age dasht overfit mishod o bad stop kon
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
#callback bayad list bashe hata baraye yki
#early_stop negah dashte
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
#dar har batch ya epoch 50 darsad neuron haro random mindaze maslan aval 15 ta

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

#layer akhar ro bayad ye activation bzarim k 0 o 1 bde mesle sigmoid
#chon k binary classification darim k khoruji 0 o 1 hast
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
#loss bayad in balayi bashe
```

```python
from tensorflow.keras.layers import Dropout
#dropout migire ye seri az neuron haro az kar mindaze
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
#in seri tulani tar shode train ma chon dropout darim 
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predictions = model.predict_classes(X_test)
#be jaye predict khali, predict_classes midim ta 0 o 1 bde
```

```python
from sklearn.metrics import classification_report,confusion_matrix
#baraye classification ina ro import mikonim
```

```python
print(classification_report(y_test,predictions))
#barresi kardan deghat
```

```python
print(confusion_matrix(y_test,predictions))
#modele ma faghat 1 dune ro miss karde baraye class 0
#va 2 ta ro baraye class 1
```
