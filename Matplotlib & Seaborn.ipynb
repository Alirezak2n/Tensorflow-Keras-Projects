```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [0,1,2]
y = [100,200,300]
plt.plot(x,y) #plot kardan khat
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
x = [0,1,2]
y = [100,200,300]
plt.plot(x,y,color="#bc67c9",marker="o",markersize=15,linestyle = '--') 
#color ezafe mikonim va marker o size marker o noe khat
plt.xlim(0,2)#had moshakhas mikonim baraye x o y
plt.ylim(0,300)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');#age nakhaym un neveshte ro bbinim ; ezafe mikonim
```
