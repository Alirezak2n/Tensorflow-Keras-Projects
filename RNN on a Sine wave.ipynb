```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,50,501)
x
```

```python
y= np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
df = pd.DataFrame(data = y,index = x,columns=['Sine'])
df#data ma shamele x o y hast k x noghate ma va y ma noghat sin ma hast ru y
```

```python
len(df)
```

```python
test_percent = 0.1
#khodemun taghsim mikonim test o train o tasmim migirim
```

```python
len(df)*test_percent
```

```python
test_point = np.round(len(df)*test_percent) #adado gerd mikone
```

```python
test_point
```

```python
test_ind = int(len(df) - test_point)
test_ind
#int mikonim k tu panda estefade konim, in cutoff mast
```

```python
train = df.iloc[:test_ind]
test = df.iloc[test_ind:]
train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#ma tu ann chon nmikhastim javab ya label ro bargardunim as input mesle rnn pas label ya 
#javab ro scale nmikardim ama inja bayad bokonim
```

```python
scaler.fit(train)
```

```python
scaled_train = scaler.transform(train)
```

```python
scaled_test = scaler.transform(test)
```

```python
#batch generator
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
help(TimeseriesGenerator)#miad baramun batch batch dorost mikone
```

```python
length = 50
#tule input hast , yani 4 ta az x axis bgir o 5omi ro predict kon ya 2ta bgir 3omi ro bgu
batch_size = 1
#chanta ro vase predict bar dare
generator = TimeseriesGenerator(data = scaled_train,targets=scaled_train,
                               length = length,batch_size=batch_size)
#data o target ma joftesh ru ye datasete yani chantasho midim baghie ro predict kon
#tu in mesal joftesh tu yek data hast
#dar vaghe meghdare x faghat sequence ro mide o bas va b dard nmikhore
```

```python
model = Sequential()

model.add(SimpleRNN(length,input_shape = (length,n_features)))

model.add(Dense(1))

model.compile(optimizer = "adam",loss='mse')
#loss mse baraye data donbale dar
```

```python
model.fit_generator(generator,epochs=5)
#chon generator estefade mikonim generator ro fit mikonim behesh
#zaman bar tare
```

```python
length = 49

generator = TimeseriesGenerator(data = scaled_train,targets=scaled_train,
                               length = length,batch_size=batch_size)

validation_generator = TimeseriesGenerator(scaled_test,scaled_test,
                                          length=length,batch_size=batch_size)
#age bare aval bzanim error mide k length andaze scaled teste k nabayad bashe
#pas length ro 49 mikonim
```
